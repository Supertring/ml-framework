# Neural Network

```python
from mllab.SimpleNN import SimpleNN

import matplotlib.pyplot as plt
import numpy as np
import sklearn
import sklearn.datasets
import sklearn.linear_model
import matplotlib
import pandas as pd

# Display plots inline and change default figure size
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
# Generate a dataset and plot it
np.random.seed(0)
X, y = sklearn.datasets.make_moons(200, noise=0.20)
X_new = np.random.rand(200,1)
X = np.append(X, X_new, axis=1)
plt.scatter(X[:,0], X[:,1], s=40, c=y, cmap=plt.cm.Spectral)
```

```python
nn = SimpleNN()
nn.add(activation="tanh",input_dim=3, hidden_dim=8, output_dim=2, print_loss=True)
nn.info()
```

```python
nn.train(X[:150], y[:150], epoches=10000, reg_value=0.01, epsilon=0.01, batch_size=32)
```

```python
pred_y = nn.infer(X[150:])
```

```python
from sklearn import metrics
print('Accuracy: %2.2f %%' % (100. * metrics.accuracy_score(y[150:], pred_y)))
```
